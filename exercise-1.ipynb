```python
percentage = df.isnull().sum()/len(df)*100
print("Missing Values Percentage Per Column:\n")
print(percentage)
```

```python
threshold = 40
dropped_columns = percentage[percentage > threshold].index
df_cleaned = df.drop(columns = dropped_columns)

print("\nColumns dropped:", list(dropped_columns))
print("\nDataset after dropping columns:\n", df_cleaned.info())
```

```python
import matplotlib.pyplot as plt
sns.histplot(df_cleaned['Booking Value'], kde=True)
plt.title("Distribution of Booking Value")
plt.show()
sns.histplot(df_cleaned['Booking Value'], kde=True)
plt.title("Distribution of Booking Value")
plt.show()
```

```python
df_cleaned['Booking Value'].fillna(df_cleaned['Booking Value'].median(), inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df_cleaned['Vehicle Type label'] = label_encoder.fit_transform(df_cleaned['Vehicle Type'])
print(df_cleaned[['Vehicle Type','Vehicle Type label']].head())
```

```python
df_one_hot = pd.get_dummies(df, columns=['Payment Method'], drop_first=True)

print("\nDataset after One-Hot Encoding:\n")
print(df_one_hot.filter(like="Payment Method").head())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_cleaned[['Booking Value','Ride Distance']] = scaler.fit_transform(df_cleaned[['Booking Value','Ride Distance']])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_cleaned[['Booking Value','Ride Distance']] = scaler.fit_transform(df_cleaned[['Booking Value','Ride Distance']])
```
